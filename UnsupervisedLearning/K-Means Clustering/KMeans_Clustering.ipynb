```python
import matplotlib.pyplot as plt
import pandas as pd 

X = pd.read_csv("dataset_movies2.csv")
X
```

```python
plt.figure()
plt.scatter(X["length"],X["costs"])
plt.show()
```

```python
from sklearn.cluster import KMeans

clust = KMeans(n_clusters=3,n_init=10).fit(X)
y = clust.predict(X) 

plt.figure()
plt.scatter(X[y==0]["length"],X[y==0]["costs"])
plt.scatter(X[y==1]["length"],X[y==1]["costs"])
plt.scatter(X[y==2]["length"],X[y==2]["costs"])
plt.show()
```

```python
# mit Z-transformierung

from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
```

```python
clust = KMeans(n_clusters=3,n_init=10).fit(X_scaled)
y = clust.predict(X_scaled) 

plt.figure()
plt.scatter(X[y==0]["length"],X[y==0]["costs"])
plt.scatter(X[y==1]["length"],X[y==1]["costs"])
plt.scatter(X[y==2]["length"],X[y==2]["costs"])
plt.show()
```

```python
clust.inertia_
```
