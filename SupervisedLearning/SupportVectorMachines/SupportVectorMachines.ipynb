```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("data/sl_svm_ex_fruits.csv")
df
```

```python
X = df.iloc[:,0:-1].values
y = df.iloc[:,2].values

plt.figure()
plt.scatter(df[y==0]["diameter"],df[y==0]["weight"], marker='x')
plt.scatter(df[y==1]["diameter"],df[y==1]["weight"], marker='s')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.inspection import DecisionBoundaryDisplay

clf = SVC(kernel='linear').fit(X,y)

plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X, plot_method="contour",levels=[-1,0,1])
plt.scatter(df[y==0]["diameter"],df[y==0]["weight"], marker='x')
plt.scatter(df[y==1]["diameter"],df[y==1]["weight"], marker='s')
plt.show()
```

```python
plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X, plot_method="contour", levels =[0])
plt.scatter(df[y==0]["diameter"],df[y==0]["weight"], marker='x')
plt.scatter(df[y==1]["diameter"],df[y==1]["weight"], marker='s')
plt.show()
```

```python
# Soft margin SVM 

df = pd.read_csv("data/sl_logreg_ex_apartment.csv")
df
```

```python
X = df.iloc[:,0:-1].values
y = df.iloc[:,2].values

plt.figure()
plt.scatter(df[y==0]["size"],df[y==0]["dist"], marker='x')
plt.scatter(df[y==1]["size"],df[y==1]["dist"], marker='s')
plt.show()
```

```python
clf = SVC(kernel='linear', C = 1).fit(X,y)

plt.figure()
DecisionBoundaryDisplay.from_estimator(clf,X, plot_method="contour",levels=[0])
plt.scatter(df[y==0]["size"],df[y==0]["dist"], marker='x')
plt.scatter(df[y==1]["size"],df[y==1]["dist"], marker='s')
plt.show()
```

```python
df = pd.read_csv("data/sl_logreg_ex_screws.csv")
df
```
